# Boxing performance tracker


## Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
print ("test")
```

## Create Dataset

```python

df = pd.read_csv("boxing_data.csv")

df

```

## Calculate Performance Metrics

```python
df["Total Rounds"] = (
    df["shadow boxing Rds"] +
    df["Heavy bag Rds"] +
    df["Body sparring Rds"] +
    df["Head sparring Rds"]
)
df["Total Sparring rounds"] = (
    df["Body sparring Rds"] +
    df["Head sparring Rds"]
)
df
```

```python
print(df.columns)
```

## Session Analysis

```python
print("Total sessions tracked:", len(df))
print("Average weight:", df["Weight"].mean())
print(
    "total sparring rounds:",
    df["Head sparring Rds"].sum() +
    df["Body sparring Rds"].sum()
)
print("Average sparring rounds:", df["Total Sparring rounds"].mean())
print("Average total rounds:", df["Total Rounds"].mean())
```

```python
df.loc[df["Total Rounds"].idxmax()]
print(df.columns)
```

```python
print("Average session duration:", df["Duration (min)"].mean())
df["Rounds per Hr"] = df["Total Rounds"]/(df["Duration (min)"]/60)
print(df[["Date", "Rounds per Hr"]]) 
```

```python
top_sessions =df.sort_values(by="Rounds per Hr", ascending=False)
print(top_sessions)
top_3 =df.sort_values(by="Rounds per Hr", ascending=False).head(3)
print(top_3)
```

### Personal Best


```python
# Most Total Rounds
best_total_rounds = df.loc[df["Total Rounds"].idxmax()]
print("🥇 Most Total Rounds")
print(best_total_rounds)
print()

# Most Sparring Rounds
best_sparring = df.loc[df["Total Sparring rounds"].idxmax()]
print("🥇 Most Sparring Rounds")
print(best_sparring)
print()

# Longest Session
longest_session = df.loc[df["Duration (min)"].idxmax()]
print("🥇 Longest Session")
print(longest_session)
print()

# Highest Rounds per Hour
highest_rounds_per_hour = df.loc[df["Rounds per Hr"].idxmax()]
print("🥇 Highest Rounds per Hour")
print(highest_rounds_per_hour)
```

## Visualisation 

```python
plt.plot(df["Date"], df["Weight"], marker="o")
plt.title("Weight over time")
plt.xlabel("date")
plt.ylabel("Weight in Kg")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.plot(df["Date"], df["Total Rounds"], marker="o")
plt.title("total rounds per session")
plt.xlabel("date")
plt.ylabel("rounds")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.plot(df["Date"], df["Duration (min)"], marker="o")
plt.title("Session Duration Over Time")
plt.xlabel("Date")
plt.ylabel("Duration (minutes)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.plot(df["Date"], df["Total Sparring rounds"], marker="o")
plt.title("Sparring Session Rounds per date")
plt.xlabel("Date")
plt.ylabel("Total Sparring rounds")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Summary

```python
weight_change = round(df["Weight"].iloc[-1] - df["Weight"].iloc[0], 2)
average_dur = round(df["Duration (min)"].mean(), 1)
average_intensity = round(df["Rounds per Hr"].mean(), 1)
highest_sparring = df.loc[df["Total Sparring rounds"].idxmax()]
total_sessions = len(df)

print(f"Weight changed by {weight_change} Kg over the tracked period.")
print(f"Average session duration was {average_dur} minute.")
print(f"Highest number of Sparring rounds in a session was {highest_sparring["Total Sparring rounds"]} rounds in one session, on {highest_sparring["Date"]}.")
print(f"Total sessions tracked: {total_sessions}.")
```
